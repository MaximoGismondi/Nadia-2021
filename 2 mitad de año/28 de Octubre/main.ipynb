```python
import pandas as pd

df_barrios = pd.read_csv("barrios.csv")
df_comunas = pd.read_csv("comunas.csv")
df_casos = pd.read_csv("casos.csv")
```

```python
condition_1 = df_barrios.comuna == 3
df_comuna_3 = df_barrios[condition_1]

print("1) En la comuna 3 se encuentran los diferentes barrios")
for i in df_comuna_3["barrio"].index:
    print(df_comuna_3["barrio"][i], "- perimetro",
          round(df_comuna_3["perimetro"][i]/100)/10, "km - area", round(df_comuna_3["area"][i]/100000)/10, "km2")
```

```python
condition_2 = (df_barrios["comuna"] > 4) & (df_barrios["comuna"] < 10)
df_comunas_entre_4_y_10 = df_barrios[condition_2]
print("2) De la comuna 5 a la 9 se encuentran los diferentes barrios")
for i in df_comunas_entre_4_y_10["barrio"].index:
    print(df_comunas_entre_4_y_10["barrio"][i], "- perimetro",
          round(df_comunas_entre_4_y_10["perimetro"][i]/100)/10, "km - area", round(df_comunas_entre_4_y_10["area"][i]/100000)/10, "km2")
```

```python
condition_3 = df_comunas["ID"] == 5
df_comuna_5 = df_comunas[condition_3]
for i in df_comuna_5["ID"].index:
    print("3) La comuna con id igual a 5 - perimetro",
          round(df_comuna_5["PERIMETRO"][i]/100)/10, "km - area", round(df_comuna_5["AREA"][i]/100000)/10, "km2")
```

```python
condition_4 = df_barrios.barrio.str.startswith('V')
df_barrios_v = df_barrios[['comuna']][condition_4]
df_barrios_v = df_barrios_v.comuna.unique()
print("4) Las comunas con barrios que comienzan con 'V'")
for i in df_barrios_v:
    print("Comuna", i)
```

```python
condition_5 = df_barrios["barrio"].str.contains("A")
df_barrios_v = df_barrios[condition_5]
print("5) Los barrios que contienen 'A'")
for i in df_barrios_v["barrio"].index:
    condition_5 = df_comunas["ID"] == df_barrios_v["comuna"][i]
    df_comunas_barrios_v = df_comunas[condition_5]
    for j in df_comunas_barrios_v["ID"].index:
        print(df_barrios_v["barrio"][i],"- comuna",df_comunas_barrios_v["ID"][j], "- area", round(df_comunas_barrios_v["AREA"][j]/100000)/10, "km2")
```

```python
condition_6 = df_comunas.ID == 15
df_comuna_id_15 = df_comunas[condition_6]

for i in df_comuna_id_15["PERIMETRO"].index:
    print("6) En la comuna con id igual a 15 - perimetro",
          round(df_comuna_id_15["PERIMETRO"][i]/100)/10, "km")
```

```python
condition_7 = df_casos["numero_de_caso"].index == 7
df_caso_indice_15 = df_casos[condition_7]
df_barrio_de_indice_15 = df_barrios[df_barrios.barrio.isin(df_caso_indice_15.barrio)]

for i in df_barrio_de_indice_15.area.index:
    print("7) El barrio con id igual a 7 - area", round(df_barrio_de_indice_15.area[i]/100000)/10, "km2")
```

```python
condition_8 = df_casos["numero_de_caso"] > 6000000
condition_8_b = df_casos["numero_de_caso"] < 7000000

df_casos_numero = df_casos[condition_8 & condition_8_b]
df_barrios_numero_casos = df_barrios[df_barrios.barrio.isin(
    df_casos_numero.barrio)]

print("8) Los barrios que tienen residenciales mayores a 6 millones y menores a 7 millones")
for i in df_barrios_numero_casos.barrio.index:
    print(df_barrios_numero_casos.barrio[i], "- indice", i,
        "- comuna", df_barrios_numero_casos.comuna[i],
        "- area", round(df_barrios_numero_casos.area[i]/100000)/10, "km2")
```

```python
condition_9 = df_barrios.index == 3
barrio_index_3 = df_barrios[condition_9]
df_casos_index_3 = df_casos[df_casos.barrio.isin(barrio_index_3.barrio)]

print("9) En el barrio con indice igual a 3 se encuentras las siguiente personas:")
for i in df_casos_index_3.index:
    print("Genero",
          df_casos_index_3["genero"][i], "-", df_casos_index_3["edad"][i], "años")
```

```python
condition_10 = df_casos.tipo_contagio == "En Investigación"
df_casos_en_investigacion = df_casos[condition_10]
df_barrios_en_investigacion = df_barrios[df_barrios.barrio.isin(df_casos_en_investigacion.barrio)]

print("10) Los barrios que tienen casos de tipo de contagio 'En Investigación'")
for i in df_barrios_en_investigacion["barrio"].index:
    condition_10_barrio = df_barrios.barrio == df_barrios_en_investigacion["barrio"][i]
    print(df_barrios_en_investigacion["barrio"][i], "- indice", i)
```

```python
condition_11 = df_casos.edad < 18
df_casos_menores_edad = df_casos[condition_11]
df_barrios_menores_edad = df_barrios[df_barrios.barrio.isin(df_casos_menores_edad.barrio)]


print("11) Los barrios que tienen casos de personas menores de edad")
for i in df_barrios_menores_edad["barrio"].index:
    condition_11_barrio = df_barrios.barrio == df_barrios_menores_edad["barrio"][i]
    print(df_barrios_menores_edad["barrio"][i], "- indice", i)
```

```python
condition_12 = df_casos.barrio.isin(df_barrios.barrio[df_barrios.index <= 4])
df_casos_barrios_no_mayores_4 = df_casos[condition_12]

print("12) Los casos que no pertenecen a los barrios con indice mayor a 4")
for i in df_casos_barrios_no_mayores_4["numero_de_caso"].index:
    print(df_casos_barrios_no_mayores_4["numero_de_caso"][i], "- indice", i, "- genero", df_casos_barrios_no_mayores_4["genero"][i], "- edad", df_casos_barrios_no_mayores_4["edad"][i])
```

```python
condition_13 = df_casos["comuna"] != 13
df_casos_comuna_distinta_de_13 = df_casos[condition_13]

print("13) Casos que no pertenecen a la comuna 13")
for i in df_casos_comuna_distinta_de_13.index:
    print(df_casos_comuna_distinta_de_13["numero_de_caso"][i], "- indice", i, "- genero", df_casos_comuna_distinta_de_13["genero"][i], "- edad", df_casos_comuna_distinta_de_13["edad"][i])
```
