```python
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.read_csv("dataset.csv")

tempMediaMax = df1.iloc[df1["media"].idxmax()]
tempMediaMin = df1.iloc[df1["media"].idxmin()]

print()
print("La temperatura media maxima fue en :", tempMediaMax.loc["mes"],"de",tempMediaMax.loc["año"] , "con",tempMediaMax.loc["media"],"C°")
print("La temperatura media minima fue en :", tempMediaMin.loc["mes"],"de",tempMediaMin.loc["año"] , "con",tempMediaMin.loc["media"],"C°")
print()
print("¿Cómo cambió la temperatura a lo largo de los años? Así:")
mediaAnual = {}
for i in df1.index:
    year = df1["año"][i]
    if year in mediaAnual.keys():
        mediaAnual[year][0] += df1["media"][i]
        mediaAnual[year][1] += 1
    else:
        mediaAnual[year] = [df1["media"][i],1]

for year in mediaAnual.keys():
    mediaAnual[year] = mediaAnual[year][0]/mediaAnual[year][1]

plt.plot( mediaAnual.keys(),mediaAnual.values())
```
