import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('mode.chained_assignment', None)

```python
df = pd.read_csv("Analisis_soporte_informatico.csv")
df2 = df[["¿Cuánto te gusta Python como lenguaje de programación?",
          "¿Hasta qué mes entregaste trabajos?", "¿Considerás que analizar datos es aburrido?"]]


df2["¿Cuánto te gusta Python como lenguaje de programación?"] = df2["¿Cuánto te gusta Python como lenguaje de programación?"].replace(
    ["Nada", "Un poco", "Normal", "Bastante", "Me encanta"], [1, 2, 3, 4, 5])
df2["¿Considerás que analizar datos es aburrido?"] = df2["¿Considerás que analizar datos es aburrido?"].replace([
                                                                                                                "Sí", "No"], [1, 2])
df2 = df2.drop(
    df2[df2["¿Cuánto te gusta Python como lenguaje de programación?"] == "me fascina"].index)

sns.pairplot(df2, hue="¿Hasta qué mes entregaste trabajos?",
             palette="Spectral", height=5)
```
