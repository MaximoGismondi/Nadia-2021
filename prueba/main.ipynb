```python
import pandas as pd
import matplotlib.pyplot as plt1
import matplotlib.pyplot as plt2
import matplotlib.pyplot as plt3
import matplotlib.pyplot as plt4

csv = pd.read_csv("dataset_testeo_turismo.csv")
x1 = csv["GRUPO_ETARIO"].unique()
y1 = csv["GRUPO_ETARIO"].value_counts().tolist()
y1.reverse()

csv = csv.loc[0:300] #sino hay muchas ubicaciones y no se lee pero anda igual
x2 = csv["FECHA_MUESTRA"].unique()
#x2 = x2.astype(str)
#x2 = x2[0:4]
#y2 = csv["FECHA_MUESTRA"].value_counts().tolist()
y2.reverse()

#grafico 1
plt1.figure(figsize=(10,5))
plt1.rcParams.update({'font.size': 15})

plt1.plot(x1,y1)
plt1.title('Testeados según grupo etario')
plt1.xlabel('Grupos etarios')
plt1.ylabel('Cantidad de testeados')
plt1.show()

plt3.figure(figsize=(10,5))
plt3.rcParams.update({'font.size': 15})

plt3.bar(x1,y1)
plt3.title('Testeados según grupo etario')
plt3.xlabel('Grupos etarios')
plt3.ylabel('Cantidad de testeados')
plt3.show()

#1. Llegamos a la conclusion de que la cantidad de testeos crece respecto a la
# edad que tengan las personas
#2. Seria conveniente saber cuantas veces se realizo el testeo a la misma persona
# para saber con exactitud el porcentaje de gente testeada segun la edad

#grafico 2
plt2.figure(figsize=(10,5))
plt2.rcParams.update({'font.size': 10})

plt2.plot(x2,y2)
plt2.title('Testeados según fecha')
plt2.xlabel('Fechas')
plt2.ylabel('Cantidad de testeados')
plt2.show()


plt4.figure(figsize=(10,5))
plt4.rcParams.update({'font.size': 10})

plt4.bar(x2,y2)
plt4.title('Testeados según ubicación')
plt4.xlabel('Fechas')
plt4.ylabel('Cantidad de testeados')
plt4.show()
```
